```python

!pip install -q pillow tensorflow scikit-learn tqdm

import os, random, json, warnings
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from PIL import Image, ImageChops, ImageEnhance, ImageDraw
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, roc_auc_score
from datetime import datetime
from google.colab import files

np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

device_name = tf.test.gpu_device_name()
print(f"✅ GPU detected: {device_name}" if device_name else " CPU mode")

print("✅ TensorFlow version:", tf.__version__)



DATASET_PATH = './dataset'
os.makedirs(f'{DATASET_PATH}/authentic', exist_ok=True)
os.makedirs(f'{DATASET_PATH}/forged', exist_ok=True)

print("\n Generating 100 synthetic demo images (50 authentic + 50 forged)")

# --- Authentic images ---
for i in range(50):
    img = Image.new('RGB', (128, 128),
                    color=(random.randint(0,255),
                           random.randint(0,255),
                           random.randint(0,255)))
    draw = ImageDraw.Draw(img)
    for _ in range(5):  # random rectangles
        xs = sorted([random.randint(0,127), random.randint(0,127)])
        ys = sorted([random.randint(0,127), random.randint(0,127)])
        draw.rectangle([xs[0], ys[0], xs[1], ys[1]], outline="white", width=1)
    img.save(f'{DATASET_PATH}/authentic/auth_{i}.jpg')

# --- Forged images (authentic + bright patches) ---
for i in range(50):
    img = Image.open(f'{DATASET_PATH}/authentic/auth_{i % 50}.jpg').copy()
    draw = ImageDraw.Draw(img)
    for _ in range(3):
        xs = sorted([random.randint(0,127), random.randint(0,127)])
        ys = sorted([random.randint(0,127), random.randint(0,127)])
        draw.rectangle([xs[0], ys[0], xs[1], ys[1]], fill="yellow")
    img.save(f'{DATASET_PATH}/forged/forg_{i}.jpg')

print("✅ Synthetic dataset created successfully!")

AUTHENTIC_DIR, FORGED_DIR = 'authentic', 'forged'

# CELL 3: ERROR LEVEL ANALYSIS (ELA)

class ELAProcessor:
    def __init__(self, quality=95):
        self.quality = quality

    def convert_to_ela(self, image_path, output_size=(128,128)):
        try:
            original = Image.open(image_path).convert('RGB')
            temp_path = f"/tmp/temp_ela_{os.path.basename(image_path)}.jpg"
            original.save(temp_path, 'JPEG', quality=self.quality)
            compressed = Image.open(temp_path)
            ela_image = ImageChops.difference(original, compressed)

            extrema = ela_image.getextrema()
            diffs = [ex[1] if isinstance(ex, tuple) else ex for ex in extrema if ex]
            max_diff = max(diffs) if diffs else 1
            if max_diff == 0: max_diff = 1

            scale = 255.0 / max_diff
            ela_image = ImageEnhance.Brightness(ela_image).enhance(scale)
            ela_image = ela_image.resize(output_size)
            os.remove(temp_path)
            return np.array(ela_image)
        except Exception as e:
            print(f"ELA Error: {e}")
            return None

ela_processor = ELAProcessor()

# CELL 4: CONVERT TO ELA DATASET

ELA_OUTPUT_PATH = './ela_dataset'
os.makedirs(f"{ELA_OUTPUT_PATH}/authentic", exist_ok=True)
os.makedirs(f"{ELA_OUTPUT_PATH}/forged", exist_ok=True)

for label in ['authentic', 'forged']:
    src = os.path.join(DATASET_PATH, label)
    print(f"⚙️ Processing {label} images for ELA...")
    for img_name in tqdm(os.listdir(src)):
        path = os.path.join(src, img_name)
        ela_img = ela_processor.convert_to_ela(path)
        if ela_img is not None:
            Image.fromarray(ela_img.astype('uint8')).save(
                f"{ELA_OUTPUT_PATH}/{label}/{img_name}")
print("ELA images ready!")


# BUILD MODEL (MobileNetV2)

def build_model():
    base = keras.applications.MobileNetV2(include_top=False,
                                          input_shape=(128,128,3),
                                          weights='imagenet')
    base.trainable = False
    model = keras.Sequential([
        base,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=keras.optimizers.Adam(1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy', keras.metrics.AUC(name='auc')])
    return model

model = build_model()
model.summary()


# DATA GENERATORS


BATCH_SIZE = 16
VAL_SPLIT = 0.2

train_datagen = ImageDataGenerator(rescale=1./255, validation_split=VAL_SPLIT,
                                   rotation_range=15, width_shift_range=0.1,
                                   height_shift_range=0.1, zoom_range=0.1,
                                   horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1./255, validation_split=VAL_SPLIT)

train_gen = train_datagen.flow_from_directory(
    ELA_OUTPUT_PATH, target_size=(128,128),
    batch_size=BATCH_SIZE, class_mode='binary',
    subset='training', shuffle=True)
val_gen = val_datagen.flow_from_directory(
    ELA_OUTPUT_PATH, target_size=(128,128),
    batch_size=BATCH_SIZE, class_mode='binary',
    subset='validation', shuffle=False)

print(f"Training samples: {train_gen.samples}, Validation: {val_gen.samples}")

#TRAIN MODEL
callbacks = [
    keras.callbacks.ModelCheckpoint('best_forgery_model.h5', save_best_only=True,
                                    monitor='val_accuracy'),
    keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                                      patience=3, verbose=1),
    keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=5,
                                  restore_best_weights=True)
]

EPOCHS = 10
print("\n Training chalu .\n")
history = model.fit(train_gen, validation_data=val_gen,
                    epochs=EPOCHS, callbacks=callbacks, verbose=1)


#EVALUATION


print("\n Evaluating model")
y_pred_prob = model.predict(val_gen)
y_pred = (y_pred_prob > 0.5).astype(int)
y_true = val_gen.classes

acc = np.mean(y_pred.flatten() == y_true)
auc = roc_auc_score(y_true, y_pred_prob)
cm = confusion_matrix(y_true, y_pred)
print(f"Accuracy: {acc*100:.2f}%, AUC: {auc:.4f}")
print("Confusion Matrix:\n", cm)

# CELL 9: VISUALIZE TRAINING

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend(); plt.title("Accuracy")
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend(); plt.title("Loss")
plt.tight_layout(); plt.show()

# CELL 10: TEST A SINGLE IMAGE

def predict_image(image_path, show_ela=True):
    ela_img = ela_processor.convert_to_ela(image_path)
    if ela_img is None: return
    ela_norm = ela_img / 255.0
    pred_prob = model.predict(np.expand_dims(ela_norm, axis=0), verbose=0)[0][0]
    label = "FORGED" if pred_prob > 0.5 else "AUTHENTIC"
    conf = pred_prob if label=="FORGED" else 1-pred_prob

    if show_ela:
        fig, ax = plt.subplots(1,2,figsize=(10,5))
        ax[0].imshow(Image.open(image_path)); ax[0].set_title("Original")
        ax[1].imshow(ela_img); ax[1].set_title(f"{label} ({conf*100:.1f}%)",
                                               color='red' if label=="FORGED" else 'green')
        plt.show()
    return {"prediction": label, "confidence": conf}

# Example prediction
test_path = f"{DATASET_PATH}/forged/forg_0.jpg"
predict_image(test_path)

# CELL 11: SAVE RESULTS


model.save('forgery_detection_model.h5')
metrics = {"accuracy": float(acc), "auc": float(auc),
           "confusion_matrix": cm.tolist(),
           "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")}
with open('final_metrics.json','w') as f: json.dump(metrics, f, indent=2)
print("Model and metrics saved!")
```
